# 1. Installation

Three options are available to install NeuRec using the command line:

* Docker image;
* Pip;
* Github repository.

> The Docker image includes Jupyter notebooks.

## Docker
The NeuRec Docker image provides a convenient and fast way to using NeuRec. CPU and GPU options are available, which can be selected by stating either 'cpu' or 'gpu' as the image, and the version number as the tag, for example '1.0.0'. The available Docker images can be [found here](https://github.com/wubinzzu/NeuRec/packages).

1. Pull the Docker image:
 
  ```bash
  docker pull docker.pkg.github.com/wubinzzu/neurec/<image>:<tag>
  ```

2. Run the image, mounting a folder of your choice into the container:

  ```bash
  docker run -p 8888:8888 -v $(pwd)/data:/tf/data docker.pkg.github.com/wubinzzu/neurec/<image>:<tag>
  ```
  
3. Connect to Juypter notebooks via a browser, and specify the token:
   
   ```
   http://127.0.0.1:8888?token=<token>
   ```
   
   > The token for Jupyter notebooks is displayed on the command line after running the image.

## Pip
A Python package is available, which can be installed using pip.

1. Setup a virtual environment to handle the packages

```bash
python3 -m venv venv

source venv/bin/activate
```

2. Install the NeuRec package:

```bash
pip3 install neurec
```

3. Import NeuRec into your project:

```python
import neurec
```

## Github
NeuRec can be installed via Github in three steps:

1. Clone the NeuRec Github repository to your local machine:

  ```bash
  git clone https://github.com/wubinzzu/NeuRec.git
  
  cd NeuRec
  ```
  
2. Setup a virtual environment to handle the packages:

    ```bash
    python3 -m venv venv
    
    source venv/bin/activate
    ```

3. Finally, install the following required python packages:

  * tensorflow 1.12.3
  * numpy 1.16.4
  * scipy 1.3.1
  * pandas 0.17
  * Cython 0.29.14
  
  ```bash
  pip3 install -r requirements.txt
  ```
  
  or
  
  ```bash
  pip3 install tensorflow==1.12.3 numpy==1.16.4 scipy==1.3.1 pandas==0.17 Cython==0.29.14
  ```
